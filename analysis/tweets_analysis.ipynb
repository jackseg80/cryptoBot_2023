```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()


endpoint = "https://api.twitter.com/2/tweets/search/recent/"
params = {
    'query': '(bitcoin OR btc) lang:en -is:retweet -has:links',
    'tweet.fields': 'created_at,lang',
    'max_results': 100,
}
headers = {"Authorization": os.environ.get("twitter_token")}

r = requests.get(endpoint, params=params, headers=headers)
req_result = dict(r.json())["data"]
tweet_df = pd.DataFrame({"tweet": [d['text'] for d in req_result]})
tweet_df
```

```python
from collections import Counter
import nltk
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('tagsets')

def group_two_dict_into_one(dict1, dict2):
    for key in dict2:
        if key in dict1.keys():
            dict1[key] += dict2[key]
        else:
            dict1["other_tag"] += dict2[key]
    return dict1

def delete_word_starting_with_in_sentence(sentence, starter):
    for word in sentence.split():
        if word.startswith(starter):
            sentence = sentence.replace(word, "")
    return sentence


str_to_test = tweet_df.iloc[2]["tweet"]
str_to_test = delete_word_starting_with_in_sentence(str_to_test, "@")
# str_to_test = "BTC: +45% today"
tokens = nltk.word_tokenize(str_to_test.lower())
text = nltk.Text(tokens)
tags = nltk.pos_tag(text)

base_value = {
    "text": str_to_test,
    "total_word": 0, 
    "total_len": 0, 
    "other_tag": 0,
    "CC": 0, 
    "CD": 0, 
    "DT": 0, 
    "EX": 0, 
    "FW": 0, 
    "IN": 0, 
    "JJ": 0, 
    "JJR": 0,
    "JJS": 0,
    "LS": 0,
    "MD": 0,
    "NN": 0,
    "NNP": 0,
    "NNPS": 0,
    "NNS": 0,
    "PDT": 0,
    "POS": 0,
    "PRP": 0,
    "PRP$": 0,
    "RB": 0,
    "RBR": 0,
    "RBS": 0,
    "RP": 0,
    "SYM": 0,
    "TO": 0,
    "UH": 0,
    "VB": 0,
    "VBD": 0,
    "VBG": 0,
    "VBN": 0,
    "VBP": 0,
    "VBZ": 0,
    "WDT": 0,
    "WP": 0,
    "WP$": 0,
    "WRB": 0
}



counts = Counter(tag for word,tag in tags)
print(str_to_test)
# counts
total = sum(counts.values())
tag_dict = dict((word, float(count)/total) for word,count in counts.items())

result_data = group_two_dict_into_one(base_value, tag_dict)
result_data["total_word"] = len(str_to_test.split())
result_data["total_len"] = len(str_to_test)
result_data
```
